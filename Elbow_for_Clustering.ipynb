```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Persiapkan data
data = np.array(list(range(1, 300))).reshape(-1, 1)

# Inisialisasi list untuk menyimpan nilai inersia
inertia = []

# Range nilai K yang ingin diuji
k_values = range(1, 11)

# Melakukan K-means clustering untuk setiap nilai K
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

# Memplot hasil Elbow method
plt.plot(k_values, inertia, 'bx-')
plt.xlabel('Jumlah K')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method')
plt.show()
```
