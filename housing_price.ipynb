```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data Training
X_train = np.array([[1000, 3, 1990],
                    [1500, 4, 1980],
                    [1200, 3, 2005],
                    [1800, 4, 1995],
                    [900, 2, 2010]])
y_train = np.array([500000, 700000, 600000, 800000, 450000])

# Inisialisasi model Regresi Linear
model = LinearRegression()

# Melatih model dengan data training
model.fit(X_train, y_train)

# Data Testing
X_test = np.array([[1300, 3, 2000],
                   [1600, 4, 1998],
                   [1100, 2, 2015]])

# Melakukan prediksi harga rumah
y_pred = model.predict(X_test)

# Visualisasi scatter plot
plt.scatter(X_train[:, 0], y_train, color='blue', label='Training Data')
plt.scatter(X_test[:, 0], y_pred, color='red', label='Predicted Data')
plt.xlabel('Luas Tanah')
plt.ylabel('Harga Rumah')
plt.legend()
plt.show()
```
